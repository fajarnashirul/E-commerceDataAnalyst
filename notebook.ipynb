## Cleaning Data

```python
# #menghapus kolom customer_unique_id pada customer_dataset
# customers_dataset_df = customers_dataset_df.drop(columns=['customer_unique_id'], axis=1)
# customers_dataset_df.head()
```

```python
#menghapus kolom geolocation_city dan geolocation_state 
geolocation_df = geolocation_df.drop(columns=['geolocation_city', 'geolocation_state'], axis=1)
#mengahapus data duplikat pada geolocation_df
geolocation_df.drop_duplicates(inplace=True)
print('data duplikat:', geolocation_df.duplicated().sum())
geolocation_df.head()
```

```python
#mengubah jenis data menjadi date_time pada orders_dataset
datetime_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

for column in datetime_columns:
    orders_dataset_df[column] = pd.to_datetime(orders_dataset_df[column])

orders_dataset_df.info()
```

```python
orders_dataset_df.describe(include='all')
```

```python
#terdapat nilai null pada order_approved_at, sehingga perlu di isi dengan mencari rata-rata waktu approve
order_approved_time = orders_dataset_df['order_approved_at'] - orders_dataset_df['order_purchase_timestamp']
order_approved_time = order_approved_time.apply(lambda x: x.total_seconds())
orders_dataset_df['order_approved_time'] = order_approved_time.round()

print('rata-rata waktu approved:', orders_dataset_df.order_approved_time.mean())

```

```python
#mengisi data kosong pada order_approved_at dengan menambah order_purchase_timestamp dengan rata-rata waktu approved
orders_dataset_df['order_approved_at'] = orders_dataset_df['order_approved_at'].fillna(
    orders_dataset_df['order_purchase_timestamp'] + pd.to_timedelta(orders_dataset_df.order_approved_time.mean().round(), unit='s'))

#mengupdate kembali order_approved_time setelah semua data terisi
orders_dataset_df['order_approved_time'] = order_approved_time.round()
orders_dataset_df.isna().sum()
```

```python
'''
pada orders_dataset_df kolom order_delivered_carrier_date dan order_delivered_customer_date,
dengan asumsi bahwa hanya order dengan status 'delivered' yang semua kolomnya perlu diisi
'''
filtered_df = orders_dataset_df[orders_dataset_df.order_status == 'delivered']
filtered_df.isna().sum()
```

```python
#pada order_delivered_customer_date yang kosong akan diisi dengan menyesuaikan order_estimated_delivered_date
orders_dataset_df.loc[orders_dataset_df['order_status'] == 'delivered', 'order_delivered_customer_date'] = orders_dataset_df.loc[orders_dataset_df['order_status'] == 'delivered', 'order_delivered_customer_date'].fillna(orders_dataset_df['order_estimated_delivery_date'])

orders_dataset_df.isna().sum()
```

```python
#melihat data order_status delivered yang memiliki order_delivered_carrier null
filtered_df = orders_dataset_df[orders_dataset_df.order_status == 'delivered']
filtered_df[filtered_df.isna().any(axis=1)]
```

```python
#nilai null pada order_delivered_carrier_date dengan status order delivered akan disi berdsarakan rata-rata waktu pengiriman
#sebelumnya akan dibuat nilai delivered_time
delivered_time = orders_dataset_df['order_delivered_customer_date'] - orders_dataset_df['order_delivered_carrier_date']
delivered_time = delivered_time.apply(lambda x: x.total_seconds())
orders_dataset_df['delivered_time'] = round(delivered_time/86400) #dikonversi kedalam hari

orders_dataset_df.describe(include='all')
```

```python
#terdapat anomali pada nilai minimum delivered time yang menghasilkan nilai minus
orders_dataset_df[orders_dataset_df['delivered_time'] < 0].head()
```

```python
#menyesuaikan order_delivered_customer_date dengan order_estimated_delivered_date agar tidak terdapat nilai minus
orders_dataset_df.loc[orders_dataset_df['delivered_time'] < 0, 'order_delivered_customer_date'] = orders_dataset_df['order_estimated_delivery_date']
#update delivered_time
delivered_time = orders_dataset_df['order_delivered_customer_date'] - orders_dataset_df['order_delivered_carrier_date']
delivered_time = delivered_time.apply(lambda x: x.total_seconds())
orders_dataset_df['delivered_time'] = round(delivered_time/86400) #dikonversi kedalam hari

orders_dataset_df.describe(include='all')
```

```python
#mengisi data kosong pada order_delivered_carrier_date dengan mengurangi order_delivered_customer_date dengan rata-rata waktu pengiriman
orders_dataset_df.loc[orders_dataset_df['order_status'] == 'delivered', 'order_delivered_carrier_date'] = orders_dataset_df.loc[orders_dataset_df['order_status'] == 'delivered', 'order_delivered_carrier_date'].fillna(
    orders_dataset_df['order_delivered_customer_date'] - pd.to_timedelta(orders_dataset_df['delivered_time'].mean().round(), unit='days')
)
#mengupdate kembali delivered_time
delivered_time = orders_dataset_df['order_delivered_customer_date'] - orders_dataset_df['order_delivered_carrier_date']
delivered_time = delivered_time.apply(lambda x: x.total_seconds())
orders_dataset_df['delivered_time'] = round(delivered_time/86400)

orders_dataset_df.isna().sum()
```

# Exploratory Data Analysis (EDA)

## Explore Data Customer

```python
#merge data customer dengan geolocation
customer_geo_df = geolocation_df.rename(columns={'geolocation_zip_code_prefix': 'customer_zip_code_prefix', 'geolocation_lat':'customer_lat', 'geolocation_lng':'customer_lng'})
customer_df = pd.merge(
    customers_dataset_df,
    customer_geo_df,
    left_on='customer_zip_code_prefix',
    right_on='customer_zip_code_prefix',
    how='left'
)

customer_df.head()
```

```python
#bersarkan data di bawah didapati bahwa customer paling banyak bersal dari kota sao paulo
customer_df.groupby(by='customer_city').customer_id.nunique().sort_values(ascending=False)
```
